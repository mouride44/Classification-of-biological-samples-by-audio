### Introduction 

Dans ce notebook , on fera une synthése sur les différentes series bonnes ou mauvaises afin d'y sortir une conclusion  et les nouvelles voies à explorer. 


Pour cela on va utiliser les bandes de fréquence harmoniques comme descripteurs et tenter de voir :

```python
# Let us load in the relevant Python modules

from pymongo import MongoClient

import matplotlib.pyplot as plt
import seaborn  as sns
import pandas as pd
import numpy as np
import plotly.plotly as py
%matplotlib inline
colormap = plt.cm.inferno
import visualisations
import warnings
from collections import Counter
from sklearn.feature_selection import mutual_info_classif
warnings.filterwarnings('ignore')
from sklearn import ensemble
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tl

#Connect in mongodb and import the entropy_ng dataset

db=MongoClient('localhost', 27017)
db=db.features_All
har = db["entropie_ng"]
#measure information 
db3=db['mesure_info']
df3 = pd.DataFrame.from_records(db3.find())

# fonction pour generer les données dans mongodb
def data_generate(reg):
    
    df = pd.DataFrame.from_records(har.find({"reglage": "C" + str(reg)}))
    df_out = df.drop(["_id", "date", "reglage"], axis=1)
    N = len(df_out.columns)
    arr_back = np.concatenate(np.concatenate(df_out.values)).reshape(N, -1).T
    df_back = pd.DataFrame(arr_back, columns=df_out.columns)
    df_back = df_back[(df_back.Classe == 1) | (df_back.Classe == 3)]
    df2 = df_back.drop_duplicates()
    df2=df2.sort_values("NumeroM")
    df2=df2.drop(["NumeroM"],axis=1)
    df2["Classe"].replace([1,3], [0, 1], inplace = True)
    y=df2.pop("Classe")
    df2 = df2.reindex_axis(sorted(df2.columns, key=lambda x: float(x[1:])), axis=1)
   
    return  df2,y


def matrix_correlation(reg):
  df2,y=data_generate(reg)
  df2 = df2.reindex_axis(sorted(df2.columns, key=lambda x: float(x[1:])), axis=1)
  plt.figure(figsize=(16,12))
  plt.title('Pearson correlation of band ', y=1.05, size=15)

  sns.heatmap(df2.iloc[:,:20].corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white',
                       annot=True)
```

```python
from xgboost import XGBClassifier

def iterative_importance(df2,y):
    gbc = XGBClassifier()
    gbc.fit(df2, y)
    feature_importance = gbc.feature_importances_
    # Scatter plot 
    trace = go.Scatter(
        y = feature_importance,
        x = df2.columns,
        mode='markers',
        marker=dict(
            sizemode = 'diameter',
            sizeref = 1,
            size = 25,
    #       size= feature_dataframe['AdaBoost feature importances'].values,
            #color = np.random.randn(500), #set color equal to a variable
            color = feature_importance,
            colorscale='Portland',
            showscale=True
        ),
        text = df2.columns
    )

    data = [trace]

    layout= go.Layout(
        autosize= True,
        title= 'Random Forest Feature Importance',
        hovermode= 'closest',
    #     xaxis= dict(
    #         title= 'Pop',
    #         ticklen= 5,
    #         zeroline= False,
    #         gridwidth= 2,
    #     ),
        yaxis=dict(
            title= 'Feature Importance',
            ticklen= 5,
            gridwidth= 2
        ),
        showlegend= False
    )
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig,filename='scatter2010')
    
```

```python
matrix_correlation(11738)
```

Ici on peut remarquer dans cette matrice de corrélation, hormis deux bandes , elles ont une corrélation assez forte . Traçons ces différentes bandes par rapport à leur valeurs d'importances c'est à dire, leurs poids sur la classification des échantillons .



```python
iterative_importance(df2,y)
```

Les bandes les plus importantes sont C1 et C16, qui correspondent aussi à celles, qui présentent une faible corrélation avec les harmonies restantes.

Cela reste cohérent avec le fonctionnement du bioscope, avec qui une modification d'un signal  est mesuré .

C'est ainsi q'une stabilité est importante , en effet si à chaque temps l'échantillon est mesuré de maniére différente  les classes peuvent chevaucher .
