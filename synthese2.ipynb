#### 1.2.1 Flux spectral 

Le flux spectral mesure la variation d'énergie entre des fréquences succéssives dans une bande.

#### 1.2.2 Roloff

Le point spectrale de Rolloff est le N% la distribution spectrale de puissance, où N est généralement de
85% où 95%. Le point de roloff est la fréquence en dessous de laquelle N% de la distribution énergétique est
concentrée. 

```python
from IPython.display import Image
Image(filename='roloff.jpg')
```

#### 1.2.3 Harmonie

Les harmonies correspondent aux multiples entiers de la fréquence fondamentale, qui est la fréquence avec la plus grande amplitude .  Pour chaque bande on determine les N harmonies , puis on fait une moyenne des harmonies qui ont la même fréquence .

```python
from IPython.display import Image
Image(filename='fft_mean.png')
```

#### 1.2.4 Bande fréquence harmonique

```python
from IPython.display import Image
Image(filename='neighbors.png')
```

Ici on peut voir la distrubition d'une bande autour d'une harmonique. Cet descripteur nous permet , au lieu de prendre seulement la valeure maximale (harmonique) , on calcule l'entropie de cette bande autour d'une harmonique .

#### 1.2.5 bande de fréquence 

Le principe est le même que celui évoqué précédemment , la différence est du fait , de ne pas utiliser seulement les bandes autour des harmonies mais d'utiliser tout le spectre . On définit une taille de bande  de 200 htz  par exemple , puis on calcule l'entropie de shannon sur chaque bande de fréquence ([0,200], [200,400], [400,600]...)

#### 1.2.6 Spectogramme

Comme on peut le remarquer , pour traiter un signal , il y'a des méthodes issues  du signal temporel et celles du signal fréquentiel  (aprés fft) . Le spectogramme nous permet de lier ces deux méthodes en combinant temps , fréquences et amplitudes. 

```python
from IPython.display import Image
Image(filename='spectogramme.png')
```

#### 1.2.7 Cohérence spectrale 

La cohérence spectrale est une statistique qui peut être utiliser pour examiner la relation entre deux signaux ou ensembles de données . Elle est couramment utilisée pour estimer le transfert de puissance entre l'entrée et la sortie d'un système linéaire . Si les signaux sont ergodiques et que la fonction du système est linéaire , elle peut être utiliser pour estimer la causalité entre l'entrée et la sortie.(source wikipedia).

Pour notre cas elle nous permettra de bien selectionner les fréquences qui sont en du  signal source et de supprimmer toutes les fréquences (bruit ,environnement...) non corrélées à ce dernier.

### 3. Distribution de différents descripteurs 

Dans cette partie , on va voir la distrubition sur quelques séries entre les différents descripteurs audio, pour ésseyer de voir leur différence . Le programme suivant  nous permet de générer, les données . 

```python
import seaborn  as sns
import pandas as pd
import numpy as np
import plotly.plotly as py
import matplotlib.pyplot as plt 
%matplotlib inline
colormap = plt.cm.inferno
import visualisations
import warnings
from collections import Counter
from sklearn.feature_selection import mutual_info_classif
warnings.filterwarnings('ignore')
from sklearn import ensemble
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tl
from pymongo import MongoClient

#Connect in mongodb and import the entropy_ng dataset

db=MongoClient('localhost', 27017)
db=db.features_All
har = db["band_freq"]
#measure information 
db3=db['mesure_info']
df3 = pd.DataFrame.from_records(db3.find())


def data_generate(reg):

    
    df = pd.DataFrame.from_records(har.find({"reglage": "C" + str(reg)}))
    df_out = df.drop(["_id", "date", "reglage"], axis=1)
    N = len(df_out.columns)
    arr_back = np.concatenate(np.concatenate(df_out.values)).reshape(N, -1).T
    df_back = pd.DataFrame(arr_back, columns=df_out.columns)
    df_back = df_back[(df_back.Classe == 1) | (df_back.Classe == 3)]
    df2 = df_back.drop_duplicates()
    df2=df2.sort_values("NumeroM")
    df2=df2.drop(["NumeroM"],axis=1)
    df2["Classe"].replace([1,3], [0, 1], inplace = True)
    
    y=df2.pop("Classe")
    df2 = df2.reindex_axis(sorted(df2.columns, key=lambda x: float(x[1:])), axis=1)
   
    return  df2,y
```

```python
har = db["entropie_ng"]
df,y=data_generate(10752)
df.shape
```

On va visualiser les distributions de quelques bandes de fréquences  entre les classes.

```python
df["Classe"]=y
g = sns.pairplot(df[["C0" ,"C1","C4","C5","C6","C20","C22","C24","Classe"]], 
                 hue='Classe',diag_kind = "kde")

def f(x, **kwargs):
    kwargs.pop("color")
    col = next(plt.gca()._get_lines.prop_cycler)['color']
    sns.kdeplot(x, color=col, **kwargs,shade=True)

g.map_diag(f)
g.map_offdiag(plt.scatter)
g.add_legend()
```

On peut remarquer,  sans utiliser des algorithmes comme les arbres de décision  pour déterminer les bandes de fréquence les plus importants, on peut voir avec seulement la distrubtion de la bande C5 par rapport aux deux classes qu'on classifie bien les échanttillons

```python

sns.FacetGrid(df, hue="Classe", size=6,).map(sns.kdeplot, "C5").add_legend()
```

Cependant , on va utiliser les  arbres de décision pour determiner, les bandes de fréquence qui expliquent au mieux les classes afin de confirmer l'hypothése enoncée précedemment .

```python
from xgboost import XGBClassifier

def iterative_importance(df2,y):
    gbc = XGBClassifier()
    gbc.fit(df2, y)
    feature_importance = gbc.feature_importances_
    # Scatter plot 
    trace = go.Scatter(
        y = feature_importance,
        x = df2.columns,
        mode='markers',
        marker=dict(
            sizemode = 'diameter',
            sizeref = 1,
            size = 25,
    #       size= feature_dataframe['AdaBoost feature importances'].values,
            #color = np.random.randn(500), #set color equal to a variable
            color = feature_importance,
            colorscale='Portland',
            showscale=True
        ),
        text = df2.columns
    )

    data = [trace]

    layout= go.Layout(
        autosize= True,
        title= 'Random Forest Feature Importance 12033',
        hovermode= 'closest',
    #     xaxis= dict(
    #         title= 'Pop',
    #         ticklen= 5,
    #         zeroline= False,
    #         gridwidth= 2,
    #     ),
        yaxis=dict(
            title= 'Feature Importance',
            ticklen= 5,
            gridwidth= 2
        ),
        showlegend= False
    )
    fig = go.Figure(data=data, layout=layout)
    py.iplot(fig,filename='scatter2010')
```

```python
df=df.drop("Classe",axis=1)
iterative_importance(df,y)
```

Comme on peut le voir, la bande C5 est celle qui contient plus d'information

Prenons une autre série, et voyons les bandes de fréquences les plus importantes

```python
har = db["band_freq"]
df,y=data_generate(12033)
iterative_importance(df,y)
```

Ici comme on peut le remarquer, les bandes les plus importantes se trouvent dans les hautes fréquences , ce qui  va dans le sens contraire de ce qu'on pensait  . Ce qui montre aussi une voie pour l'exploration des données audios comme image et d'analyser les axes temps , fréquence et amplitude (spectogramme) .
